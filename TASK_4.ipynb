```python
import pandas as pd
df = pd.read_csv('adult.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['education'] = le.fit_transform(df['education'])
```

```python
df = pd.get_dummies(df, columns=['marital-status', 'relationship', 'race', 'gender'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
num_cols = ['age', 'capital-gain', 'hours-per-week']
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
# This shows the first few rows of your new, encoded, and scaled data
print("Processed Data Preview:")
print(df.head())

# This confirms the new data types and that scaling worked
print("\nData Summary:")
print(df.describe())
```

```python
#  Save processed dataset
df.to_csv('processed_adult_income.csv', index=False)
print("File saved successfully as 'processed_adult_income.csv'")
```
